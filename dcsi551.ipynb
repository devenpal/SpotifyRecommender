Search Dataset by Name of Song

```python
output = widgets.Output()

ALL = 'ALL'

def sorted_vals(array):
    unique_val = array.unique().tolist()
    unique_val.sort()
    unique_val.insert(0, ALL)
    return unique_val

song_search = widgets.Dropdown(options = sorted_vals(subset.name), description = 'Name: ')

def song_search_eventhandler(change):
    output.clear_output()
    if (change.new == ALL):
        display(subset)
    else:
        display(subset[subset.name == change.new])


song_search.observe(song_search_eventhandler, names='value')
display(song_search)
```

```python
#loudness, tempo, popularity

@interact
def display_user_input_features(column=['danceability', 'energy', 'speechiness', 'acousticness', 'liveness', 'instrumentalness', 'valence'], 
                            x=(0, 1, 0.01)):
  return subset.loc[subset[column] > x]
```

Machine Learning Portion:

```python
from pyspark import SparkContext
from pyspark.sql import SparkSession
from pyspark.ml.feature import BucketedRandomProjectionLSH, VectorAssembler
from pyspark.ml.linalg import Vectors
from pyspark.sql.types import DoubleType
import pyspark.sql.functions as f

spark = SparkSession.builder.getOrCreate()


df = spark.read.csv('/content/gdrive/My Drive/tracks_features.csv', header=True)

try:
  var1 = input("Enter attribute 1: ")

  var2 = input("Enter attribute 2: ")

  attr_df = df.select(f.col("id"),f.col("name"),f.col("artists"),f.col(var1),f.col(var2))

  import pyspark.sql.functions as f

  attr_df = attr_df.withColumn(var1, f.col(var1).cast('double'))
  attr_df = attr_df.withColumn(var2, f.col(var2).cast('double'))

  assembler = VectorAssembler(
      inputCols=[var1, var2],
      outputCol="features",
      handleInvalid = "skip")

  feat_df = assembler.transform(attr_df)

  brp = BucketedRandomProjectionLSH(inputCol="features", outputCol="hashes", bucketLength=2.0,
                                    numHashTables=3)
  model = brp.fit(feat_df)

  song = input("Enter song name: ")

  song_row = attr_df.filter("name = " + song).limit(1).collect()
  song_attr1 = song_row[0][var1]
  song_attr2 = song_row[0][var2]

  key = Vectors.dense([song_attr1, song_attr2])

  recommended = model.approxNearestNeighbors(feat_df, key, 11).collect()

  print("We recommend these songs: ")
  for song_row in recommended:
    if song_row["name"] != song:
      artists = song_row["artists"].strip("[]").replace("'","")
      print(song_row["name"] + " by " + artists)

except:
  print("Please try again with valid attributes or song name.")




```
